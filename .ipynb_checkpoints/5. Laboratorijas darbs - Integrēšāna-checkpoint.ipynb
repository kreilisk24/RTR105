## Integrēšana ar laukuma palīdzību

### Funkcija cos(x^2)



Funkcijas integrāli, reķinot ar **WolframAlpha**, iegūst šādu vērtību:


```python
from math import cos
from numpy import random
from matplotlib import pyplot as plt
```

Iestātisim mainīgos:

```python
N = 3000 # Punktu skaits
x0 = 0   # x ass intervaala saakums
x1 = 4   # intervaal beigas
y0 = -1  # y ass intervaala saakums
y1 = 1   # intervaala beigas
```

Ar skaitļu ģeneratoru radīsim skaitļu masīvu, kas ar *uniform* metodi vienmērīgi "izsvaida" punktus funkcijas robežās.

```python
x = random.uniform(x0,x1,N)
y = random.uniform(y0,y1,N)
```

Definēsim doto funkciju:

```python
def f(x):
    return cos(x**2)
```

#### Grafika attēlošana

```python
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funkcija un taas integraalis (laukums starp funkciju un x ass)')
#Iestata mainīgo, kas skaita punktus funkcijas laukumā
N1 = 0
for i in range(N):
# Ja punkta vērtība ir lielāka par x asi un mazāka par funkcijas vērtību,
# Tad punktus attēloju zaļā krāsā,
# Pēc integrāļa definīcijas šis laukums ir ar pozitīvu zīmi,
# Tātad N1 punktu skaitītājam pievienoju šo punktu 
    if y[i] >= 0 and y[i] <= f(x[i]) :
        plt.plot(x[i],y[i],'go')
        N1 = N1 + 1
# Ja punkta vērtība ir mazāka par x asi, bet lielāka par funkcijas vērtību,
# Tad punktus attēloju zilā krāsā,
# Šie punkti atrodas laukumā ar negatīvo zīmi,
# Tāpēc no N1 punktu skaitītāja atņemu šo punktu
    elif y[i] <= 0 and y[i] >= f(x[i]):
        plt.plot(x[i],y[i],'bo')
        N1 = N1 - 1
# Pārējie punkti mums nav svarīgi, tāpēc attelosim tos sarkanus un N1 nemainām
    else:
        plt.plot(x[i],y[i],'ro')
plt.show()
```

Tā kā mums ir nepieciešams izrēķināt funkcijas laukumu, tad izveidojam vienādības:

```python
S_zinaamais = (x1-x0) * (y1-y0)
S_nezinaamais = (1. * S_zinaamais * N1)/ N
# Viss grafika laukums reizināts ar vajadzīgo punktu skaitu,
# Rezultātā dod funkcijas laukuma punktu skaitu, ko dalot ar visiem punktiem
# Iegūst integrāli cos(x^2), kur x pieder intervālam [0;4]
print(S_nezinaamais)
```

Laukums ir tuvu dotajam **WolframAlpha** vērtībai, bet nav precīzs, tāpēc ka, tas ir atkarīgs no skaitļu ģeneratora izvēlētajiem punktiem un to skaitu.  
(*Ar 5000 skaitļiem esmu ieguvis vērtību 6.0096*)
