## Atvasināšana 

### Funkcija cos(x^2)

```python
from numpy import cos, linspace
from matplotlib import pyplot as plt
```

Definēsim mūsu funkciju un atlasīsim 100 tās punktus intervālā (-4;4)

```python
def f(x):
    return cos(x*x)
x = linspace(-4,4,100)
y = f(x)
```

Grafika veidošanai izmantojam *pyplot* funkcijas  
Ar *plt.plot* uzzīmēsim pirmo grafiku - cos(x^2)

```python
legend = []
plt.axis([-4,4, -10, 10])
plt.grid()
plt.xlabel('x')
plt.ylabel('cos(x^2)')
plt.title('Funkcijas $cos(x^2) un taas atvasinaajumi')
plt.plot(x, y,"k")
legend.append("$cos(x^2)$")
plt.plot(x, y,"ro")
legend.append("cos(x^2)$ - punkti")
```

Ar *desmos* grafiku zīmētāju var pierādīt, ka grafiks ir pareizs:


Lai iegūtu atvasinājumu ir nepieciešams izveidot tās funkciju:


Atlasīsim punktu skaitu un definējam *delta*x

```python
N = len(x)
deltax = x[1] - x[0]
```

Tā kā atvasinājumam zīmēsim grafiku, tam nepieciešams atvēlēt skaitļu masīvu

```python
atvasinajums = []
```

Lai iegūtu skaitļu masīvu, ar *for* ciklu ievietojam atvasinājuma funkciju

```python
for i  in range(N):
    temp = (f(x[i] + deltax) - f(x[i])) / deltax
#rezultātu ievieto skaitļu masīvā
    atvasinajums.append(temp)
```

Tālāk to var ievietot koordinātu asīs

```python
plt.plot(x,atvasinajums,"y")
legend.append("$cos(x^2)$ atvasinaajums")
plt.plot(x,atvasinajums,"mo")
legend.append("$cos(x^2)$ atvasinajums - punkti")
plt.show()
```

Ar *desmos* pārbaudam funkciju -2xsin(x^2), kas ir cos(x^2) atvasinājums:


Lai iegūtu 2. kārtas atvasinājumu, 1. kārtas atvasinājumu ievietosim tajā pašā *for* ciklā, bet šoreiz *cos(x^2)* vietā ir pirmās kārtas atvasinājums

```python
def f2(x):
    return (f(x + deltax) - f(x)) / deltax
atvas2 = []
for i in range(N):
    temp2 = (f2(x[i] + deltax) - f2(x[i])) / deltax
    atvas2.append(temp2)
```

Attēlosim koordinātu asīs

```python
plt.plot(x,atvas2,"g")
legend.append("$cos(x^2)$ - 2. kartas atvasinaajums")
plt.plot(x,atvas2,"bo")
legend.append("$cos(x^2)$ - 2.k atvasinajuma punkti")
```

2.kārtas atvasinājums cos(x^2) = 
*desmos* grafiks:


Visi grafiki vienās koordinātu asīs:

```python
legend = []
plt.axis([-4,4, -10, 10])
plt.grid()
plt.xlabel('x')
plt.ylabel('cos(x^2)')
plt.title('Funkcijas $cos(x^2) un taas atvasinaajumi')
plt.plot(x, y,"k")
legend.append("$cos(x^2)$")
plt.plot(x, y,"ro")
legend.append("$cos(x^2)$ - punkti")
plt.plot(x,atvasinajums,"y")
legend.append("$cos(x^2)$ atvasinaajums")
plt.plot(x,atvasinajums,"mo")
legend.append("$cos(x^2)$ atvasinajums - punkti")
plt.plot(x,atvas2,"g")
legend.append("$cos(x^2)$ - 2. kartas atvasinaajums")
plt.plot(x,atvas2,"bo")
legend.append("$cos(x^2)$ - 2.k atvasinajuma punkti")
plt.legend(legend, loc = 3,fancybox=True, framealpha =0.55, facecolor="white")
plt.show()
```
