## Zīmēšana ar *pyplot*
### Funkcija *sin(x)*

#### 1. Lai veidotu grafikus, vispirms nepiciešams izveidot skaitļu masīvus ar kuru palīdzību tiek zīmēta funkcija koordinātu asī  
Lai darbotos ar zīmēšanas darbarīkiekiem python vidē, no *numpy* nepieciešams importēt skaitļu ģeneratoru

```python
from numpy import *
```

Lai iegūtu skaitļu masīvu izmanto *linspace(intervāla sākums, beigas, skaitļu daudzums)*

```python
x = linspace(0,4,70)
print(x)
```

Skaitļu masīvam izrēķinam sinusus

```python
y = sin(x)
```

Pirmo grafiku apzīmēsim kā *f1*, kuram pamatā ir funkcija y = x

```python
f1 = x
```

Izveidosim palīgdarbības, kas nākamos grafikus pietuvinās reālajai *sin* funkcijai, pēc Teilora rindas: 


```python
f = float(1*2*3)
s = (x**3/f)
```

Otro grafiku *f2* veidosim pēc Teilora rindas pirmajiem diviem locekļiem

```python
f2 = f1 - s
```

Trešo grafiku *f3* veido, pievienojot iepriekšējam grafikam *f2* nākamo Teilora rindas locekli

```python
f3 = f2 + (x**5)/(f*4*5)
```

Ceturto grafiku *f4*, veido pēc tā paša principa (nākamais Teilora rindas loceklis)

```python
f4 = f3 - (x**7)/(f*4*5*6*7)
```

#### 2. Grafika veidošana ar *pyplot* no *matplotlib* bibliotēkas

```python
from matplotlib import pyplot as plt
```

Uzstādam koordinātu asis un tās apzīmē  
Ar *plt.plot* uzstādam *y = sin(x)* grafiku

```python
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Funkcija $sin(x)$")
plt.plot(x, y, color = "#000000", label="sin")
```

Tālāk pievienojam pārējos grafikus

```python
plt.plot(x, f1, label="f1")
plt.plot(x, f2, label="f2")
plt.plot(x, f3, label="f3")
plt.plot(x, f4, label="f4")
```

Lai redzētu iegūtos grafikus, pieprasam *pyplot*  
*plt.show()*  
Ar *plt.legend()* visus grafikus, kam ir *label=*, attēlojam leģendu apzīmējumos

```python
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funkcija $sin(x)$")
plt.plot(x, y, color = "#000000", label="sin")
plt.plot(x, f1, label="f1")
plt.plot(x, f2, label="f2")
plt.plot(x, f3, label="f3")
plt.plot(x, f4, label="f4")
plt.legend()
plt.show()
```

Redzams, ka, jo vairāk locekļu ir izmantoti no Teilora rindas, jo tuvāks kļūst grafiks reālajam *sin(x)* grafikam
