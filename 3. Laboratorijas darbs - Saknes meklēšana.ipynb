## Saknes meklēšana ar dihotomijas metodi

### Funkcija cos(x^2)

1. No *math* bibliotēkas atlasīsim nepieciešamās funkcijas  
*cos* - funkcijas apreķināšanai  
*fabs(x)* - iegūst absolūto x vērtību 

```python
from math import cos, fabs
```

Definēsim savu funkciju, lai tā nav jāraksta katru reizi

```python
def f(x):
    return cos(x*x)
```

Tad izvēlamies funkcijas grafika intervālu, kur ir sakne:  
Izmantojot *gnuplot* funkciju
Ievadot: plot [0:4] cos(x*x)
Iegūst šādu attēlu:

                          1. gnuplot attēls

Attēlā redzams, ka sakne atrodas apmēram pie x = 1.25

Tātad es izvēlos intervālu starp 1.1 un 1.7

```python
a = 1.1
b = 1.7
```

Lai dihotomijas metode strādātu, ar python to var pārbaudīt šādi:

```python
funa = f(a)
funb = f(b)
```

```python
# Ja iegūto punktu reizinājums ir pozitīvs, tas nozīmē,
# ka abi punkti ir vai nu zem x ass, vai virs x ass un sakņu nav
if ( funa * funb > 0.0 ):
    print("Dotajaa intervala saknu nav", a, b)
else:
    print("Dotaja intervala saknes(s) ir!")
```

Iestatām mainīgos, kas palīdzēs noteikt, cik reizes jāsamazina intervāls, lai iegūtu sakni ar precizitāti - 0.0001  
k - iterācijas skaitītājs

```python
deltax = 0.0001
k = 0
```

#### Sašaurināšana
Lai sašaurinātu intervālu līdz *deltax*, izmantosim ciklu:

```python
while (fabs(float(a)-float(b)) > deltax):
    k = k + 1
# Dala intervālu uz pusēm
    x = (float(a)+float(b))/2
# Iegūtajam punktam izrēķina cos(x*x)
    funx = f(x)
# Lai zinātu, kura intervāla daļa ir lieka,
# Rezultātu reizina ar intervāla sākumu,
# Salīdzinot ar 0, ja rezultāts negatīvs,
# tad sakne ir pa kreisi no intervāla viduspunkta,
# Tātad intervālu sašaurina no labās puses(b = x)
    if (funa*funx < 0. ):
        b = x
# Rezultāts pozitīvs, tātad sakne ir pa labi, sašaurina no kreisās puses
    else:
        a = x
```

Kad cikls ir pabeigts var pieprasīt nepieciešamās vērtības.

```python
print("Funkcijas saknes vertiba: x = %.4f "%x)
print("Dotas funkcijas vertiba saja punkta: cos(x^2) = %f "%f(x))
print("Nepieciesamo iteraciju skaits intervalu dalisanai uz pusem: k = ",k)
```
